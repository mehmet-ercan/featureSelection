```python
import mne
import pandas as pd
import numpy as np
```

```python

raw = mne.io.read_raw_edf('dataset/Pt1_S1.edf', preload=True)
```

```python
raw.filter(l_freq=1.0, h_freq=40.0)
```

```python
features = pd.DataFrame()

for channel in raw.ch_names:
    data = raw.get_data(picks=channel)

    if data.ndim > 1:
        data = data.flatten()

    features[channel] = data

print(features.head())
```
